## 🚀 Execução Sequencial

Executa cada notebook em ordem, capturando métricas e tratando erros.

```python
# Variáveis para tracking
results = []
total_start_time = time.time()

print("🚀 Iniciando execução sequencial...")
print()

for idx, notebook in enumerate(notebooks, 1):
    print("="*70)
    print(f"📓 [{idx}/{len(notebooks)}] Executando: {notebook['name']}")
    print(f"   Descrição: {notebook['description']}")
    print(f"   Path: {notebook['path']}")
    print()
    
    # Marca início
    nb_start_time = time.time()
    success = False
    error_message = None
    result_value = None
    
    try:
        # Executa o notebook
        result_value = dbutils.notebook.run(
            notebook['path'],
            timeout_seconds=notebook['timeout']
        )
        
        # Calcula tempo
        nb_duration = int(time.time() - nb_start_time)
        
        # Verifica resultado
        if result_value and result_value.upper() in ["NO_NEW_DATA", "NO_DEPUTADOS"]:
            print(f"   ℹ️  Resultado: {result_value}")
            print(f"   ⏱️  Tempo: {nb_duration}s")
            success = True
        else:
            print(f"   ✅ Concluído com sucesso!")
            print(f"   ⏱️  Tempo: {nb_duration}s")
            success = True
        
        # Registra resultado
        results.append({
            "notebook": notebook['name'],
            "status": "SUCCESS",
            "duration_sec": nb_duration,
            "result": result_value,
            "error": None
        })
        
    except Exception as e:
        # Erro na execução
        nb_duration = int(time.time() - nb_start_time)
        error_message = str(e)
        
        print(f"   ❌ FALHA: {error_message}")
        print(f"   ⏱️  Tempo até falha: {nb_duration}s")
        
        # Registra erro
        results.append({
            "notebook": notebook['name'],
            "status": "FAILED",
            "duration_sec": nb_duration,
            "result": None,
            "error": error_message
        })
        
        # Para a execução
        print()
        print("="*70)
        print("🛑 ABORTANDO TESTE - Falha detectada")
        print("="*70)
        break
    
    print()

# Calcula tempo total
total_duration = int(time.time() - total_start_time)

print("="*70)
print("🏁 Execução concluída")
print(f"⏱️  Tempo total: {total_duration}s ({total_duration/60:.1f} minutos)")
print("="*70)
print()
```
